# Predicting on the test dataset 

## Loading the necessary libraries

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
```

We load the data in the same way that we used to generate for training the dataset

```python
import os
test_img=[]
test_label=[]
image_size=224
test_path=r"/kaggle/input/brain-tumor-mri-dataset/Testing"

model = load_model('/kaggle/input/mriiiiiiiiiii/model3_weights.h5')


label_list=os.listdir(test_path)
label_list
for x in label_list:
    
    for file in os.listdir(os.path.join(test_path,x)):
        file_path=os.path.join(test_path,x,file)
        if(x=='notumor'):
            test_label.append(0)
        else:
            test_label.append(1)
        # Load the image to predict
        img = load_img(file_path, target_size=(64, 64), color_mode='grayscale')

        # Convert the image to a numpy array
        img_arr = img_to_array(img)

        # Expand the dimensions of the image to match the input shape of the model
        img_arr = img_arr.reshape((1,) + img_arr.shape)

        # Preprocess the image using the generator used for training
        generator = ImageDataGenerator(rescale=1./255)
        img_arr = generator.standardize(img_arr)
        test_img.append(img_arr)
```

This is how the test_img looks, we have loaded all the images into the list at once

## Now we shuffle the test_img and test_label so that we get a clear idea

```python
from sklearn.utils import shuffle
```

```python
test_img, test_label = shuffle(test_img, test_label, random_state=0)
```

```python
test_label
```

```python
y_pred=[]
for i in test_img:
    prediction = model.predict(i)

    # Print the predicted class
    predicted_class = prediction.argmax()    
    if predicted_class==2:
        y_pred.append(0)
    else:
        y_pred.append(1)
```

```python
y_pred
```

# Making a confusion matrix for True Positive, True Negative, False Positive, False Negative 

```python
from sklearn.metrics import confusion_matrix

# Create confusion matrix
cm = confusion_matrix(test_label, y_pred)

print(cm)
```

# Calculating the F1 score, precision and recall

```python
from sklearn.metrics import classification_report

report = classification_report(test_label, y_pred)
print(report)
```
